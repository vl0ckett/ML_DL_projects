# Softmax in code

```python
import numpy as np
```

```python
a = np.random.randn(5)
a
```

```python
expa = np.exp(a)
expa
```

```python
answer = expa/np.sum(expa)
answer
```

```python
total_prob = np.sum(answer)
total_prob
```

```python
A = np.random.randn(100, 5)  # 100 samples, 5 classes
A
```

```python
expA = np.exp(A)
expA
```

```python
answer = expA / expA.sum(axis=1, keepdims=True)
answer
```

```python
answer.sum(axis=1)
```

```python
prediction_labels = np.argmax(answer, axis=1)
prediction_labels
```
