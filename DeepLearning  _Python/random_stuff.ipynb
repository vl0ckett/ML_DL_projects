```python
import numpy as np


def softmax(a):
    expa = np.exp(a)
    return expa / expa.sum(axis=1, keepdims=True)


X = np.array([[1, 2]])
W1 = np.array([[1, 1], [1, 0]])
W2 = np.array([[0, 1], [1, 1]])
Z = np.tanh(X.dot(W1))
out = softmax(Z.dot(W2))

print(f"{out}")
```
